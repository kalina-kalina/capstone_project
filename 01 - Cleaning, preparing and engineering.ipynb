## 4. Feature engineering 

The 'accords' column contains a few labels for each row, e.g.:

```python
#checking construction of 'accords' column
df['accords'][0] 
```

Cleaning this column:
1. Spliting each accord into separeted columns
2. Removing interrupted labels - 'Videos' and 'Pictures'

These two labels are the result of web scrapping. In the majory of cases they are located in the last two positions. Therefore, the last two columns are worthless. When a perfume has a lower amount of accords, 'Videos' and 'Pictures' can also be in one of previous columns.

3. Changing columns names into names that are easier to understand

```python
df = pd.concat([df,df.accords.str.split(',',expand=True)],1) #splitting 'accords' into separated columns
```

```python
df.head(2) # Now, separated accords are in columns 0,1,2,3,4,5,6 and 7
```

```python
#checking what is in column '6'
df[6].unique()
```

```python
#In columns '6' and '7' are worthless labes, there are here because of mistake in web scparring
#I decicded to remove them
df = df.drop([6,7], axis = 1)
```

```python
#changing new columns' names
df = df.rename(columns={0: 'accord_1', 1: 'accord_2', 2: 'accord_3', 3:'accord_4', 4: 'accord_5', 5: 'accord_6'})
```

```python
df.head(2)
```

```python
#removing worthless labels from others columns
df = df.replace({'Pictures': None})
```

```python
df = df.replace({'Videos': None})
```

Creating a set of unique accords:

```python
#one list of lists from columns: 'accord_1' - 'accord_6'
accords_from_all_columns = list(df['accord_1'].unique()) + list(df['accord_2'].unique()) + list(df['accord_3'].unique()) + list(df['accord_4'].unique()) + list(df['accord_5'].unique())  + list(df['accord_6'].unique()) 
```

```python
#removing None values
cleaned_accords = list(filter(None, accords_from_all_columns))
```

```python
list_of_accords = np.array(cleaned_accords)
```

```python
unique_accords = np.unique(list_of_accords) #list of unique accords' labels
```

```python
len(unique_accords) # -> in the dataset there are 76 different accords' labels
```

```python
df.shape # -> now 35 columns
```

Creating new columns for each of the accords and filling them with 0 (when a perfume doesn't have this accord) and 1 (when a perfume has this accord):

```python
#loop to create empty column for each accord's label
for i in unique_accords:
    df[i] = 0.0
```

```python
#if a perfume has specific accord, filling the cell with 1
for i in range(0, df.shape[0]):
    for acc in ('accord_1', 'accord_2', 'accord_3', 'accord_4', 'accord_5', 'accord_6'):
        p = df[acc][i]
        if p != None:
              df[p][i] = df[p][i] +1
```

```python
df.shape # -> After adding unique accords' columns -> 111 columns in the dataset
```

```python
df.head(2)
```

Creating a new variable 'longevity':

The new variable is a weighted aritmetical mean, with weights:

* 1 for longevity_poor
* 2 for longevity_weak
* 3 for longevity_moderate
* 4 for longevity_long
* 5 for longevity_very_long

```python
df['longevity'] = np.nan
```

```python
# calculating weighted aritmetical mean for longevity
for i in range (0, df.shape[0]):
    k = (df['longevity_poor'][i] + df['longevity_weak'][i] + df['longevity_moderate'][i] + df['longevity_long'][i] + df['longevity_very_long'][i])
    if k != 0 : 
        df['longevity'][i] = round(((df['longevity_poor'][i]*1 + df['longevity_weak'][i]*2 + df['longevity_moderate'][i]*3 + df['longevity_long'][i]*4 + df['longevity_very_long'][i]*5)/k),2)
    else:
        df['longevity'][i] = 0.0
```

```python
df['longevity'].hist(bins=50, xlabelsize = 14, ylabelsize = 14, figsize = (14,8))
plt.title("Distribution of longevity variable", size = 16)
plt.xlabel("'longevity'", size = 16)
plt.ylabel("Frequency", size = 16)
```

Creating a new variable 'sillage':

The new variable is a weighted aritmetical mean, with weights:

* 1 for sillage_soft
* 2 for sillage_moderate
* 3 for sillage_heavy
* 4 for sillage_enormous

```python
df['sillage'] = np.nan
```

```python
# calculating weighted aritmetical mean for sillage
for i in range(0, df.shape[0]):
    m = (df['sillage_soft'][i] + df['sillage_moderate'][i] + df['sillage_heavy'][i] + df['sillage_enormous'][i])
    if m != 0 :
        df['sillage'][i] = round(((df['sillage_soft'][i]*1 + df['sillage_moderate'][i]*2 + df['sillage_heavy'][i]*3 + df['sillage_enormous'][i]*4)/m),2)
    else:
        df['sillage'][i] = 0.0
```

```python
df.head(2)
```

```python
df['sillage'].hist(bins=50, xlabelsize = 14, ylabelsize = 14, figsize = (14,8))
plt.title("Distribution of sillage variable", size = 16)
plt.xlabel("'sillage'", size = 16)
plt.ylabel("Frequency", size = 16)
```

Adding conditions to eliminate incorect rows:

```python
df = df[df['longevity'] >= 1.0]
df = df[df['sillage'] >= 1.0]
```

```python
df = df.reset_index(drop=True)
```

Creating new columns to see what percentage of users voted for each season:

```python
# new empty columns
df['spring'] = np.nan
df['summer'] = np.nan
df['autumn'] = np.nan
df['winter'] = np.nan
```

```python
# filling with 0.0
for i in ['winter','summer','spring','autumn'] :
    df[i] = 0.0
```

```python
# calculating percentage of users voted for each season
for i in range (0, df.shape[0]):
    r = df['clswinter'][i] + df['clsspring'][i] + df['clssummer'][i] + df['clsautumn'][i]
    if r != 0:
        df['winter'][i] = round(df['clswinter'][i]/r,2)
        df['spring'][i] = round(df['clsspring'][i]/r,2)
        df['summer'][i] = round(df['clssummer'][i]/r,2)
        df['autumn'][i] = round(df['clsautumn'][i]/r,2)
```

```python
df.head(2)
```
