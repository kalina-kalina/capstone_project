The dataset includes 86 columns and 51,212 rows, but lots of them are worthless.

Variables have different types of values. Some of the features are categorical, some of them are numerical.

Columns meaning:

* 'brand', 'title', 'date' - basic informations about a perfume
* 'accords' - main elements/ingredients of a fragrance
* 'rating_score' - mean value of users votes
* 'votes' - the number of votes submitted by users
* 'longevity_poor', ... , 'longevity_very_long' - description of perfume's longevity 
* 'sillage_soft', ... , 'sillage_enormous' - description of perfume's sillage
* I.'clslove', 'clslike', 'clsdislike' - number of clicks on the love/like/dislike button
* II.'clswinter', 'clsspring', 'clssummer','clsautumn' - number of clicks on the winter/spring/summer/autumn button
* III. 'clscold', 'clshot' - number of clicks on the cold/hot button
* IV. 'clsday', 'clsnight' - number of clicks on the day/night button
* 'clslove_female25under', ... , 'clsnight_male25older' - I. : IV. split into sex and age categories
* 'Ihaveit', 'Ihadit', 'Iwantit' - number of clicks on the "I have/had/want it" button
* 'notes_1', ... , 'notes_20' - all elements/ingredients of a fragrance
* 'gender' - who is the perfume dedicated for (women/man/unisex)


## 3. Data cleaning

Checking for duplicated rows:

```python
df_all.duplicated().sum()
```

```python
# Coś o tym dlaczego usuwam <20 votes
```

```python
df_all['rating_score'].hist(bins=70)
```

Distribiuton of 'rating_score' variable is disrupted by products with a low amout of votes.

The less votes, the more likely the value of 'rating_score' is the integer number.

Creating a new dataframe without them to get a distribution close to the normal distribution:

```python
more20 = df_all[df_all['votes'] >= 20]
```

```python
more20['rating_score'].hist(bins=70)
```

```python
more20.shape
```

Dropping columns not used in modelling:

```python
df_no_age = more20.drop(['clslove_female25under', 'clslove_male25under', 'clslove_female25older',
       'clslove_male25older', 'clslike_female25under', 'clslike_male25under',
       'clslike_female25older', 'clslike_male25older',
       'clsdislike_female25under', 'clsdislike_male25under',
       'clsdislike_female25older', 'clsdislike_male25older',
       'clswinter_female25under', 'clswinter_male25under',
       'clswinter_female25older', 'clswinter_male25older',
       'clsspring_female25under', 'clsspring_male25under',
       'clsspring_female25older', 'clsspring_male25older',
       'clssummer_female25under', 'clssummer_male25under',
       'clssummer_female25older', 'clssummer_male25older',
       'clsautumn_female25under', 'clsautumn_male25under',
       'clsautumn_female25older', 'clsautumn_male25older',
       'clsday_female25under', 'clsday_male25under', 'clsday_female25older',
       'clsday_male25older', 'clsnight_female25under', 'clsnight_male25under',
       'clsnight_female25older', 'clsnight_male25older'], axis = 1)
```

```python
df = df_no_age.drop(['date','notes_1', 'notes_2', 'notes_3', 'notes_4',
       'notes_5', 'notes_6', 'notes_7', 'notes_8', 'notes_9', 'notes_10',
       'notes_11', 'notes_12', 'notes_13', 'notes_14', 'notes_15', 'notes_16',
       'notes_17', 'notes_18', 'notes_19', 'notes_20'], axis = 1)
```

Looking for null values:

```python
df.isna().sum()
```

```python
df['gender'].unique()
```

Replacing Nan values with 'unisex' label:

```python
df['gender'].fillna('unisex', inplace=True)
```

Dropping rows without values in 'accords' column:

```python
df = df.dropna()
```

Resetting the index column of the dataframe:

```python
df = df.reset_index(drop=True)
```

The 'accords' column contains a few labels for each row, e.g.:

```python
df['accords'][0]
```

Cleaning this column:
1. Spliting each accord into separeted columns
2. Removing interrupted labels - 'Videos' and 'Pictures'

These two labels are the result of web scrapping. In the majory of cases they are located in the last two positions. Therefore, the last two columns are worthless. When a perfume has a lower amount of accords, 'Videos' and 'Pictures' can also be in one of previous columns.

3. Changing columns names into names that are easier to understand

```python
df = pd.concat([df,df.accords.str.split(',',expand=True)],1)
```

```python
df.head(2)
```

```python
df[7].unique()
```

```python
df[6].unique()
```

```python
df = df.drop([6,7], axis = 1)
```

```python
df = df.rename(columns={0: 'accord_1', 1: 'accord_2', 2: 'accord_3', 3:'accord_4', 4: 'accord_5', 5: 'accord_6'})
```

```python
df.head(2)
```

```python
df = df.replace({'Pictures': None})
```

```python
df = df.replace({'Videos': None})
```

Creating a set of unique accords:

```python
accords_from_all_columns = list(df['accord_1'].unique()) + list(df['accord_2'].unique()) + list(df['accord_3'].unique()) + list(df['accord_4'].unique()) + list(df['accord_5'].unique())  + list(df['accord_6'].unique()) 
```

```python
cleaned_accords = list(filter(None, accords_from_all_columns))
```

```python
list_of_accords = np.array(cleaned_accords)
```

```python
unique_accords = np.unique(list_of_accords)
```

```python
len(unique_accords)
```

```python
df.shape
```

Creating new columns for each of the accords and filling them with 0 (when a perfume doesn't have this accord) and 1 (when a perfume has this accord):

```python
for i in unique_accords:
    df[i] = 0.0
```

```python
for i in range(0, df.shape[0]):
    for acc in ('accord_1', 'accord_2', 'accord_3', 'accord_4', 'accord_5', 'accord_6'):
        p = df[acc][i]
        if p != None:
              df[p][i] = df[p][i] +1
```

```python
df.shape
```

```python
df.head(2)
```

Creating a new variable 'longevity':

The new variable is a weighted aritmetical mean, with weights:

* 1 for longevity_poor
* 2 for longevity_weak
* 3 for longevity_moderate
* 4 for longevity_long
* 5 for longevity_very_long

```python
df['longevity'] = np.nan
```

```python
for i in range (0, df.shape[0]):
    k = (df['longevity_poor'][i] + df['longevity_weak'][i] + df['longevity_moderate'][i] + df['longevity_long'][i] + df['longevity_very_long'][i])
    if k != 0 : 
        df['longevity'][i] = round(((df['longevity_poor'][i]*1 + df['longevity_weak'][i]*2 + df['longevity_moderate'][i]*3 + df['longevity_long'][i]*4 + df['longevity_very_long'][i]*5)/k),2)
    else:
        df['longevity'][i] = 0.0
```

```python
df['longevity'].hist(bins=50)
```

Creating a new variable 'sillage':

The new variable is a weighted aritmetical mean, with weights:

* 1 for sillage_soft
* 2 for sillage_moderate
* 3 for sillage_heavy
* 4 for sillage_enormous

```python
df['sillage'] = np.nan
```

```python
for i in range(0, df.shape[0]):
    m = (df['sillage_soft'][i] + df['sillage_moderate'][i] + df['sillage_heavy'][i] + df['sillage_enormous'][i])
    if m != 0 :
        df['sillage'][i] = round(((df['sillage_soft'][i]*1 + df['sillage_moderate'][i]*2 + df['sillage_heavy'][i]*3 + df['sillage_enormous'][i]*4)/m),2)
    else:
        df['sillage'][i] = 0.0
```

```python
df.head(2)
```

```python
df['sillage'].hist(bins=50)
```

Adding conditions to eliminate incorect rows:

```python
df = df[df['longevity'] >= 1.0]
df = df[df['sillage'] >= 1.0]
```

```python
df = df.reset_index(drop=True)
```

Creating new columns to see what percentage of users voted for each season:

```python
df['spring'] = np.nan
df['summer'] = np.nan
df['autumn'] = np.nan
df['winter'] = np.nan
```

```python
for i in ['winter','summer','spring','autumn'] :
    df[i] = 0.0
```

```python
for i in range (0, df.shape[0]):
    r = df['clswinter'][i] + df['clsspring'][i] + df['clssummer'][i] + df['clsautumn'][i]
    if r != 0:
        df['winter'][i] = round(df['clswinter'][i]/r,2)
        df['spring'][i] = round(df['clsspring'][i]/r,2)
        df['summer'][i] = round(df['clssummer'][i]/r,2)
        df['autumn'][i] = round(df['clsautumn'][i]/r,2)
```

```python
df.head(2)
```

Changing the 'gender' column from categorical into boolean:

```python
df = pd.get_dummies(data=df, columns=['gender'])
```

```python
df.head(2)
```

Dropping some columns to avoid duplicated features:

```python
df.columns.tolist()
```

```python
df_base = df.drop(['accords', 
                   'longevity_poor', 'longevity_weak', 'longevity_moderate', 'longevity_long', 'longevity_very_long',
                   'sillage_soft', 'sillage_moderate', 'sillage_heavy', 'sillage_enormous',
                   'clswinter', 'clsspring', 'clssummer', 'clsautumn',
                   'accord_1', 'accord_2', 'accord_3', 'accord_4', 'accord_5', 'accord_6'], axis = 1)
```

And some others which will be not usefull for modelling:

```python
df_base = df_base.drop(['brand', 'title',
                   'clslove', 'clslike', 'clsdislike',
                   'clscold', 'clshot',
                   'clsday', 'clsnight',
                   'Ihaveit', 'Ihadit', 'Iwantit',], axis = 1)
```

Setting the same data type for the whole dataframe:

```python
df_base = df_base.astype(float)
```

## 4. Data saving

Lastly, saving the data as its own cleaned .csv file to be used in modelling:
